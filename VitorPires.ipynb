```python
df = pd.read_sql_query(query+" ORDER by u.id asc, e.id asc;", con2)
# Criar os widgets
buscar_input = pn.widgets.TextInput(name='BUSCAR', placeholder='Digite aqui')
metodo_radio = pn.widgets.RadioButtonGroup(
    name='Selecione método',
    options=['ID', 'Nome', 'Bairro','Descri'],
    button_type='primary'
)
id_usuario_input = pn.widgets.TextInput(name='ID usuario', placeholder='Digite aqui')
id_end_input = pn.widgets.TextInput(name='ID end', placeholder='Digite seu endereco')
rua_input = pn.widgets.TextInput(name='Rua', placeholder='Digite aqui')
bairro_input = pn.widgets.TextInput(name='Bairro', placeholder='Digite aqui')
numero_input = pn.widgets.TextInput(name='Numero', placeholder='Digite aqui')
cep_input = pn.widgets.TextInput(name='CEP', placeholder='Digite aqui')
desc_input = pn.widgets.TextInput(name='Descricao', placeholder='Digite a Descricao')

buscar_button = pn.widgets.Button(name='Buscar', button_type ='primary')
inserir_button = pn.widgets.Button(name='Inserir', button_type='primary')
excluir_button = pn.widgets.Button(name='Excluir', button_type='primary')
atualizar_button = pn.widgets.Button(name='Atualizar', button_type='primary')
limpar_button = pn.widgets.Button(name='Limpar', button_type='primary')
# Criar a tabela
table = pn.widgets.Tabulator(df,sortable = False,formatters={"numero": {"type": "number", "decimals": 0},"id_end": {"type": "number", "decimals": 0}},disabled=True, height=200)
# Funções para os botões (você pode implementar a lógica posteriormente)
import panel as pn
import pandas as pd

# Supondo que `buscar_input` seja um TextInput e `metodo_radio` seja um RadioButtonGroup
buscar_input = pn.widgets.TextInput(name="Buscar", placeholder="Insira o valor")
metodo_radio = pn.widgets.RadioButtonGroup(name="Método", options=["ID", "Nome", "Bairro","Descri"])

# Função para realizar a busca
def buscar(event):
    global df
    # Obtém o valor do filtro e remove espaços extras
    filtro = buscar_input.value.strip().lower()
    metodo = metodo_radio.value

    try:
        # Verifica se o filtro está vazio
        if not filtro:
            df = pd.read_sql_query(query+" ORDER by u.id asc, e.id asc;", con2)
            table.value = df
            return pn.pane.Alert("O campo de busca não pode estar vazio.", alert_type="warning")

        # Constrói a consulta SQL com base no método selecionado
        if metodo == "ID":
            # Verifica se o filtro é um número inteiro
            if not filtro.isdigit():
                return pn.pane.Alert("O ID deve ser um número inteiro.", alert_type="warning")
            query2 = query + " WHERE u.id = %s"
            params = (int(filtro),)
        elif metodo == "Nome":
            query2 = query + " WHERE u.nome ILIKE %s"
            params = (f"%{filtro}%",)
        elif metodo == "Bairro":
            query2 = query + " WHERE e.bairro ILIKE %s"
            params = (f"%{filtro}%",)
        elif metodo == "Descri":
            query2 = query + " WHERE ed.descricao ILIKE %s"
            params = (f"%{filtro}%",)
        else:
            return pn.pane.Alert("Método de busca inválido.", alert_type="danger")

        # Executa a consulta com o filtro
        df_filtrado = pd.read_sql_query(query2 + " ORDER BY u.id ASC, e.id ASC;", con2, params=params)

        # Verifica se o DataFrame filtrado está vazio
        if df_filtrado.empty:
            return pn.pane.Alert("Nenhum resultado encontrado.", alert_type="info")
        df = df_filtrado
        # Atualiza a tabela com os resultados filtrados
        table.value = df
        buscar_input.value = ""
    except Exception as e:
        # Retorna um alerta em caso de erro
        buscar_input.value = ""
        return pn.pane.Alert(f"Erro ao realizar a busca: {e}", alert_type="danger")

def inserir(event):
    global df
    cursor = con.cursor()
    try:
        # Verifica se algum dos valores é None ou vazio
        if any(value is None or value == "" for value in [rua_input.value, cep_input.value, numero_input.value, bairro_input.value, id_usuario_input.value,desc_input.value]):
            print("Erro: Todos os campos devem ser preenchidos.")
            raise ValueError("Um ou mais campos estão vazios ou nulos.")  # Lança uma exceção para interromper o fluxo
    
        # Cria o cursor
        cursor = con.cursor()
    
        # Comando SQL para inserir dados na tabela Endereco
        insert_endereco_query = '''
            INSERT INTO Endereco (Rua, CEP, Numero, Bairro)
            VALUES (%s, %s, %s, %s)
            RETURNING id;  -- Retorna o ID do endereço inserido
        '''
    
        # Dados a serem inseridos na tabela Endereco
        novo_end = (rua_input.value, cep_input.value, numero_input.value, bairro_input.value)
    
        # Executa o comando INSERT e obtém o ID do endereço inserido
        cursor.execute(insert_endereco_query, novo_end)
        endereco_id = cursor.fetchone()[0]  # Recupera o ID do endereço
    
        # Comando SQL para inserir o relacionamento na tabela endereco_de
        insert_endereco_de_query = '''
            INSERT INTO endereco_de (id_usuario, id_endereco, descricao)
            VALUES (%s, %s, %s)
        '''
    
        # Dados a serem inseridos na tabela endereco_de
        novo_relacionamento = (id_usuario_input.value, endereco_id, desc_input.value)
    
        # Executa o comando INSERT para criar o relacionamento
        cursor.execute(insert_endereco_de_query, novo_relacionamento)
    
        # Confirma as alterações no banco de dados
        con.commit()
        print("Dados inseridos com sucesso!")
    
    except ValueError as ve:
        # Captura a exceção específica para valores nulos
        print(f"Erro de validação: {ve}")
    except Exception as e:
        # Captura qualquer outro erro
        print(f"Erro ao inserir dados: {e}")
        if con:
            con.rollback()  # Desfaz as alterações em caso de erro
    finally:
        # Fecha o cursor e a conexão
        df = pd.read_sql_query(query+" ORDER by u.id asc, e.id asc;", con2)
        table.value = df
        limpar(event)
        if cursor:
            cursor.close()
            
def excluir(event):
    global df
    cursor = con.cursor()
    try:
        # Verifica se algum dos valores é None ou vazio
        if any(value is None or value == "" for value in [id_end_input.value]):
            print("Erro: O campo Id end devem estar preenchido para uma exclusão.")
            raise ValueError("Campo nescessario para exclusao esta vazio")  # Lança uma exceção para interromper o fluxo
        # Cria o cursor
        check_id_query = '''
            SELECT COUNT(*) FROM endereco WHERE id = %s;
        '''
        cursor.execute(check_id_query, (id_end_input.value,))
        result = cursor.fetchone()[0]  # Obtém o resultado da contagem

        if result == 0:
            raise ValueError(f"O ID Endereço {id_end_input.value} não existe no banco de dados.")
            
        
        # Exclui o relacionamento na tabela endereco_de
        delete_relacionamento_query = '''
            DELETE FROM endereco_de
            WHERE id_endereco = %s
        '''
        cursor.execute(delete_relacionamento_query, (id_end_input.value,))
    
        # Exclui o endereço na tabela endereco
        delete_endereco_query = '''
            DELETE FROM endereco
            WHERE id = %s
        '''
        cursor.execute(delete_endereco_query, (id_end_input.value,))
        # Confirma as alterações no banco de dados
        con.commit()
        
        print("Endereço excluído com sucesso!")
    
    except ValueError as ve:
        # Captura a exceção específica para valores nulos ou registros não encontrados
        print(f"Erro de validação: {ve}")
    except Exception as e:
        # Captura qualquer outro erro
        print(f"Erro ao excluir dados: {e}")
        if con:
            con.rollback()  # Desfaz as alterações em caso de erro
    finally:
        # Fecha o cursor e a conexão
        df = pd.read_sql_query(query+" ORDER by u.id asc, e.id asc;", con2)
        table.value = df
        limpar(event)
        if cursor:
            cursor.close()
    

def atualizar(event):
    global df
    cursor = con.cursor()
    # Inicializa as variáveis fora do bloco try para evitar UnboundLocalError
    try:
        # Verifica se algum dos valores é None ou vazio
        if any(value is None or value == "" for value in [id_end_input.value,rua_input.value, cep_input.value, numero_input.value, bairro_input.value, id_usuario_input.value, desc_input.value]):
            print("Erro: Todos os campos devem ser preenchidos.")
            raise ValueError("Um ou mais campos estão vazios ou nulos.")  # Lança uma exceção para interromper o fluxo

        
        # Verifica se o id_endereco existe na tabela endereco
        check_id_query = '''
            SELECT COUNT(*) FROM endereco WHERE id = %s;
        '''
        cursor.execute(check_id_query, (id_end_input.value,))
        result = cursor.fetchone()[0]  # Obtém o resultado da contagem

        if result == 0:
            raise ValueError(f"O ID Endereço {id_end_input.value} não existe no banco de dados.")

        query_update = '''
            UPDATE endereco
            SET numero = %s, cep = %s, rua = %s, bairro = %s
            WHERE id = %s;
        '''
        cursor.execute(query_update, (
            numero_input.value,
            cep_input.value,
            rua_input.value,
            bairro_input.value,
            id_end_input.value
        ))
        query_r_update = '''
                        UPDATE endereco_de
                        SET descricao = %s
                        WHERE id_endereco = %s;
        '''
        cursor.execute(query_r_update, (
            desc_input.value,
            id_end_input.value
        ))
        # Confirma as alterações no banco de dados
        con.commit()

        print("Dados atualizados com sucesso!")

    except ValueError as ve:
        # Captura erros de validação
        print(f"Erro de validação: {ve}")
    except Exception as e:
        # Captura qualquer outro erro
        print(f"Erro ao executar operação: {e}")
        if con:
            con.rollback()  # Desfaz as alterações em caso de erro
    finally:
        # Fecha o cursor e a conexão, verificando se eles foram inicializados
        df = pd.read_sql_query(query+" ORDER by u.id asc, e.id asc;", con2)
        table.value = df
        limpar(event)
        if cursor:
            cursor.close()

def limpar(event):
    id_usuario_input.value = ""
    id_end_input.value = ""
    rua_input.value = ""
    bairro_input.value = ""
    numero_input.value = ""
    cep_input.value = ""
    desc_input.value = ""

    
def select_on_click(event):
    if event.row is not None:
        selected_row_index = event.row  # Obtém o índice da linha clicada
        selected_row = df.iloc[selected_row_index]  # Obtém a linha selecionada do DataFrame
        
        # Preenche os campos com base nos valores da linha selecionada
        id_usuario_input.value = str(selected_row['id'])
        
        # Função auxiliar para tratar valores nulos ou NaN
        def safe_value(value):
            if pd.isna(value):  # Verifica se o valor é NaN
                return ""
            elif isinstance(value, (int, float)):  # Se for número, remove casas decimais
                return str(int(value))
            elif value is None:  # Verifica se o valor é None
                return ""
            else:
                return str(value)  # Converte para string
        id_end_input.value = safe_value(selected_row['id_end'])
        rua_input.value = safe_value(selected_row['rua'])
        bairro_input.value = safe_value(selected_row['bairro'])
        numero_input.value = safe_value(selected_row['numero'])  # Trata o campo 'numero'
        cep_input.value = safe_value(selected_row['cep'])
        desc_input.value = safe_value(selected_row['descricao'])
        
table.on_click(select_on_click)

# Associar as funções aos botões
inserir_button.on_click(inserir)
excluir_button.on_click(excluir)
atualizar_button.on_click(atualizar)
buscar_button.on_click(buscar)
limpar_button.on_click(limpar)
# Layout da interface
layout = pn.Column(
    pn.Row(
        pn.Column(
            
            id_usuario_input,
            id_end_input,
            rua_input,
            bairro_input,
            numero_input,
            cep_input,
            desc_input,
            pn.Row(inserir_button, excluir_button, atualizar_button,limpar_button),
            buscar_input,
            pn.Row(metodo_radio, buscar_button)
        ),
        pn.Column(
            pn.pane.Markdown('## Tabela'),
            table
        )
    ),
)
pn.pane.Alert('## Alert\nThis is a warning!')
# Mostrar a interface
layout.servable()
```
